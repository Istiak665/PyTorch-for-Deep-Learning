# Now let's use a instance method for more easy accessable for instance variable

```python
class AboutStudent:
  def __init__(self, name, Id):
    self.name = name
    self.id = Id

  # Instance Method
  def Details(self):
    print("Student Name:", self.name, "ID:", self.id)
```

```python
# Let's Create Object
StudentA = AboutStudent("Rahim", 1001)
StudentB = AboutStudent("Karim", 2001)
```

```python
# Let's print memory location
print(StudentA)
print(StudentB)
```

```python
# Let's print objects
StudentA.Details()
StudentB.Details()
```

```python
# Let's update StudentA ID
StudentA.id = 3001
# Let's print StudentA Details
StudentA.Details()
```

## Example 2 - House Design

```python
class House:
  def __init__(self):
    self.window = 4
    self.door = 2

  def view(self):
    print(self.window, "Windows", self.door, "Doors")
```

```python
# Create House object
House1 = House()
House2 = House()
```

```python
# Print details of House1 and House2
House1.view()
House2.view()
```

```python
# We can update. suppose we need 6 windows of House1 and 10 Windows of House2
House1.window = 6
House2.window = 10
```

```python
# Print update of details of House1 and House2
House1.view()
House2.view()
```

## Example 3 - Car design

```python
class Car:
  def __init__(self, name, model):

    # Instance variables
    self.name = name
    self.model_year = model
    self.wheel = 4

  # Instance method
  def view(self):
    print("The model year of this", self.name, "is", self.model_year)
    print("It is a", self.wheel, "car")
```

```python
# Create object
carA = Car("BMW", 2016)
```

```python
# Print carA details
carA.view()
```

```python
# Create another car object
carB = Car("Audi", 2018)
```

```python
# Print carA details
carB.view()
```

## Example 4- Dog

```python
class Dog:
  def __init__(self, name, color):

    # Instance variable(Attributes)
    self.name = name
    self.color = color

  # Instance Method
  # Suppose we want to change color from user so we create a method for that
  def update_color(self, color):
    self.color = color

  # Another Instance method
  # We want to view
  def poke(self):
    print("This dog name is", self.name, "and it's color",self.color)
```

```python
# Let's create dog object
dogA = Dog("Rover", "Brown")
dogB = Dog("Tom", "Black")
dogC = Dog("Lalu", "White")
```

```python
# Let's print memory location for all dog object
print(dogA)
print(dogB)
print(dogC)
```

```python
# Let's print details of all dog objects
dogA.poke()
dogB.poke()
dogC.poke()
```

```python
# Now we want to update each dogA color
dogA.update_color("Orange")
dogB.update_color("Red")
dogC.update_color("Yellow")
```

```python
# Now print update color of each dog
dogA.poke()
dogB.poke()
dogC.poke()
```

## Example 5 - Book

```python
class Book:
  def __init__(self, name, author):
    # Instance variable or attributes
    self.name = name
    self.author = author
    self.price = 0

  # Instance method for price update of book. Set method update korbe
  def set_price(self, price):
    self.price = price

  # Another Instance method for show update price of book to the user. It also called get method
  def get_price(self):
    return self.price

  # Another method for details
  def details(self):
    print(f"Book Name: {self.name}")
    print(f"Author Name: {self.author}")
    print(f"Book Price (BDT): {self.price}")
```

```python
# Create object
book1 = Book("Opekkha", "Humayan Ahamed")
```

```python
# Let's print details
book1.details()
```

* **Now we wnt to update price**

```python
# Let's update this book price
book1.set_price(200)
```

* **Now check whether it update or not**

```python
# Let's check the updated price
book1.get_price()
```

```python
# Let's now print all information of book1 including update price
book1.details()
```

```python
# Let's print all attribute as dictionary
print(book1.__dict__)
```

```python
# Let's check all attributes
print(dir(book1))
```

## Example - Cat object with compare method

```python
class Cat:
  def __init__(self, color, action):

    #Instance variables
    self.color = color
    self.action = action

  # Instance method
  def view(self):
    print(self.color, "Cat is", self.action)

  # Instance method
  def compare(self, another_cat):
    if self.action == another_cat.action:
      print("Both Cat are", self.action)
    else:
      print("Their action are different")
```

```python
# Create Cat objects
catA = Cat("Black", "Sleeping")
catB = Cat("White", "Eating")
catC =Cat("Brown", "Sleeping")
```

```python
# View Details
catA.view()
catB.view()
catC.view()
```

```python
# Let's compare (ref_object.compare())
# Let's compare catA and catB. Here ref object catA
catA.compare(catB)
```

```python
# Let's compare catA and catC. Here ref object catA
catA.compare(catC)
```
