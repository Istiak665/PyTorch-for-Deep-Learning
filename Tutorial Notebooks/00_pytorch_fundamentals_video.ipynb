#Matrix multiplication

Two main ways of performing multiplication in neural networks and deep learning:
1. Element-wise multiplication
2. Matrix multiplication (Dot product)

One of the most common operations in machine learning and deep learning algorithms (like neural networks) is matrix multiplication.

PyTorch implements matrix multiplication functionality in the `torch.matmul()` method.

The main two rules for matrix multiplication to remember are:

1. The **inner dimensions** must match:
* `(3, 2) @ (3, 2)` won't work
* `(2, 3) @ (3, 2)` will work
* `(3, 2) @ (2, 3)` will work

2. The resulting matrix has the shape of the **outer dimensions**:
* `(2, 3) @ (3, 2)` -> `(2, 2)`
* `(3, 2) @ (2, 3)` -> `(3, 3)`

Note: `"@"` in Python is the symbol for matrix multiplication.

```python
# Let's check
# first rule
# multiply (3, 2) @ (3, 2)
# It will not work an error will create
torch.matmul(torch.rand(3,2), torch.rand(3,2))
```

```python
# Let's check
# first and second rule
# multiply (2, 3) @ (3, 2)
# The resulting matrix has the shape of the outer dimensions:
# Output dimension (2, 2)
torch.matmul(torch.rand(2,3), torch.rand(3,2))
```

```python
# Let's check
# first and second rule
# multiply (3, 2) @ (2, 3)
# The resulting matrix has the shape of the outer dimensions:
# Output dimension (3, 3)
torch.matmul(torch.rand(3,2), torch.rand(2,3))
```

# One of the most common errors in deep learning (shape errors)

Because much of deep learning is multiplying and performing operations on matrices and matrices have a strict rule about what shapes and sizes can be combined, one of the most common errors you'll run into in deep learning is shape mismatches.

```python
# Shapes need to be in the right way  
tensor_C = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)

tensor_D = torch.tensor([[7, 10],
                         [8, 11], 
                         [9, 12]], dtype=torch.float32)
```

```python
# Matrix multiplication
torch.matmul(tensor_C, tensor_D) # (this will error)
```

We can make matrix multiplication work between `tensor_C` and `tensor_D` by making their inner dimensions match.

One of the ways to do this is with a **transpose** (switch the dimensions of a given tensor).

You can perform transposes in PyTorch using either:

* `torch.transpose(input, dim0, dim1)` - where input is the desired tensor to transpose and `dim0` and `dim1` are the dimensions to be swapped.
`tensor.T` - where tensor is the desired tensor to transpose.

```python
# View tensor_C and tensor_D
print(tensor_C)
print(tensor_D)
```

```python
# View tensor_C and tensor_D.T
print(tensor_C)
print(tensor_D.T)
```

```python
# The operation works when tensor_D is transposed
print(f"Original shapes: tensor_C = {tensor_C.shape}, tensor_D = {tensor_D.shape}\n")
print(f"New shapes: tensor_C = {tensor_C.shape} (same as above), tensor_D.T = {tensor_D.T.shape}\n")
print(f"Multiplying: {tensor_C.shape} * {tensor_D.T.shape} <- inner dimensions match\n")
print("Output:\n")
output = torch.matmul(tensor_C, tensor_D.T)
print(output) 
print(f"\nOutput shape: {output.shape}")
```

You can also use `torch.mm()` which is a short for torch.matmul(). And Without the transpose, the rules of matrix mulitplication aren't fulfilled and we get an error like above.

You can create your own matrix multiplication visuals like this at http://matrixmultiplication.xyz/

```python
# torch.mm is a shortcut for matmul
torch.mm(tensor_C, tensor_D.T)
```

# Change tensor datatype

As mentioned, a common issue with deep learning operations is having your tensors in different datatypes.

If one tensor is in `torch.float64` and another is in `torch.float32`, you might run into some errors.

But there's a fix.

You can change the datatypes of tensors using `torch.Tensor.type(dtype=None)` where the dtype parameter is the datatype you'd like to use.

First we'll create a tensor and check it's datatype (the default is `torch.float32`).

```python
# Create a tensor and check its datatype
tensor5 = torch.arange(10., 100., 10.)
tensor5.dtype
```

Now we'll create another tensor the same as before but change its datatype to `torch.float16`

```python
# Create a float16 tensor
tensor_float16 = tensor5.type(torch.float16)
tensor_float16
```

And we can do something similar to make a `torch.int8` tensor.

```python
# Create a int8 tensor
tensor_int8 = tensor5.type(torch.int8)
tensor_int8
```

Note: Different datatypes can be confusing to begin with. But think of it like this, the lower the number (e.g. 32, 16, 8), the less precise a computer stores the value. And with a lower amount of storage, this generally results in faster computation and a smaller overall model. Mobile-based neural networks often operate with 8-bit integers, smaller and faster to run but less accurate than their float32 counterparts.

# squeeze and unsqueeze

How about removing all single dimensions from a tensor?

To do so you can use `torch.squeeze()` (I remember this as squeezing the tensor to only have dimensions over 1).

```python
# Let's create a tensor

example1 = torch.arange(1, 8)
example1, example1.shape
```

```python
# Reshape above tensor
example1_reshape = example1.reshape(1, 7)
example1_reshape, example1_reshape.shape
```

```python
# squeeze
# Remove extra dimension from x_reshaped by using Squeeze
example1_squeezed = example1_reshape.squeeze()
print(f"\nNew tensor: {example1_squeezed}")
print(f"New shape: {example1_squeezed.shape}")
```

And to do the reverse of `torch.squeeze()` you can use `torch.unsqueeze()` to add a dimension value of 1 at a specific index.

```python
# Add a dimension value of 1 at a specific index use unsqueeze

print(f"Previous tensor: {example1_squeezed}")
print(f"Previous shape: {example1_squeezed.shape}")

## Add an extra dimension with unsqueeze
example1_unsqueezed = example1_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {example1_unsqueezed}")
print(f"New shape: {example1_unsqueezed.shape}")
```

```python
# Add a dimension value of 1 at a specific index use unsqueeze

print(f"Previous tensor: {example1_squeezed}")
print(f"Previous shape: {example1_squeezed.shape}")

## Add an extra dimension with unsqueeze
example1_unsqueezed = example1_squeezed.unsqueeze(dim=1)
print(f"\nNew tensor: {example1_unsqueezed}")
print(f"New shape: {example1_unsqueezed.shape}")
```

# Permute

You can also rearrange the order of axes values with `torch.permute(input, dims)`, where the input gets turned into a view with new dims.

```python
# Create tensor with specific shape
x_original = torch.rand(size=(224, 224, 3)) # [height, width, color_channels]

# Permute the original tensor to rearrange the axis order
# Let's change the dimension order [color_channels, height, width] -> (3, 224, 224)
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
```

## Indexing (selecting data from tensors)

Sometimes you'll want to select specific data from tensors (for example, only the first column or second row).

To do so, you can use indexing.

If you've ever done indexing on Python lists or NumPy arrays, indexing in PyTorch with tensors is very similar.

```python
# Create a tensor 
import torch
x = torch.arange(1, 10).reshape(1, 3, 3)
x, x.shape
```

Indexing values goes outer dimension -> inner dimension (check out the square brackets).

```python
# Let's index on above tensor(First bracket)
x[0]
```

```python
# Let's index on middle bracket(second bracket) (dim=1)
x[0][0]
```

```python
# Print all
print(f"{x[0][0]}")
print(f"{x[0][1]}")
print(f"{x[0][2]}")
```

```python
# Let's index on the most inner bracket (last dimension)
x[0][0][0]
```

```python
# Let's index all elements on the most inner bracket (last dimension)
print(f"innner 0-0 index: {x[0][0][0]}")
print(f"innner 0-1 index: {x[0][0][1]}")
print(f"innner 0-2 index: {x[0][0][2]}")

# 2nd row
print(f"innner 0-0 index: {x[0][1][0]}")
print(f"innner 0-1 index: {x[0][1][1]}")
print(f"innner 0-2 index: {x[0][1][2]}")

# 3rd row
print(f"innner 0-0 index: {x[0][2][0]}")
print(f"innner 0-1 index: {x[0][2][1]}")
print(f"innner 0-2 index: {x[0][2][2]}")
```

# PyTorch tensors & NumPy

Since NumPy is a popular Python numerical computing library, PyTorch has functionality to interact with it nicely.

The two main methods you'll want to use for NumPy to PyTorch (and back again) are:

* `torch.from_numpy(ndarray)` - NumPy array -> PyTorch tensor. (**Numpy array to PyTorch Tensor**)
* `torch.Tensor.numpy()` - PyTorch tensor -> NumPy array. (**Pytorch tensor to Numpy Array**)

```python
# Numpy Array to Tensor

import torch
import numpy as np

np_array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(np_array)
np_array, tensor
```

Here we can see that the tensor data type is float64. Why?
Because the **default data type of numpy array** is `float64`. But in general **tensor default data type** is `float32`. But we can change by below process

```python
tensor_32 = torch.from_numpy(np_array).type(torch.float32)
tensor_32, tensor_32.dtype
```

* **What happen with tensor if we change the values of numpy array? Let's try and check**

```python
# Add 1 with all elements of array
np_array = np_array + 1

# Now check new array and tensor
np_array, tensor
```

* **New tensor doesn't change if you change the original array.**

```python
# Now convert tensor to numpy array

new_tensor = torch.ones(7)
numpy_array = new_tensor.numpy()

new_tensor, numpy_array
```

```python
# Chcek data type
print(f"Tensor data type: {new_tensor.dtype}")
print(f"Converted numpy array data type: {numpy_array.dtype}")
```

## Reproducbility (Trying to take random out of random)

In short how a neural network learns:

`Start with random numbers -> tensor operations -> Update the random numbers to try and make them better representations of the data -> again ->again -> again---`

* To reduce the randomness in neural network and PyTorch come the concept of a **random seed**

Essentially what the random seed does is "flavour" the randomness

```python
import torch

# Create twon random tensor

random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A==random_tensor_B)
```

```python
# Let's make some random but reproducible tensors
import torch

# Set the random seed
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)

# Let's vreate random tensor
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C==random_tensor_D)
```
